```python
dataset.drop(['hash','last_refresh','id','disaster_closeout_date','incident_end_date','declaration_date','incident_begin_date'],axis = 1 , inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
LE = LabelEncoder ()
dataset['fema_declaration_string'] = LE.fit_transform(dataset['fema_declaration_string'])
dataset['state'] = LE.fit_transform(dataset['state'])
dataset['declaration_type'] = LE.fit_transform(dataset['declaration_type'])
dataset['incident_type'] = LE.fit_transform(dataset['incident_type'])
dataset['declaration_title'] = LE.fit_transform(dataset['declaration_title'])
dataset['designated_area'] = LE.fit_transform(dataset['designated_area'])
```

```python
dataset
```

```python
sns.heatmap(dataset)
```

```python
sns.pairplot(dataset)
```

```python
y = dataset.iloc[:,3].values
y
```

```python
x = dataset.iloc[:,:]
```

```python
x.drop(['declaration_type'], axis=1, inplace=True)
```

```python
x.values
```

```python
sns.boxplot(data=x)
```

```python
sns.catplot(data=x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x , y , test_size = 0.2 , random_state = 0 )
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(x_train,y_train)
y_pred = svc.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
df_comp = pd.DataFrame({'Actual':y_test , 'Predict':y_pred})
df_comp
```
